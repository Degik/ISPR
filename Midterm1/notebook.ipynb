# Read Dataset

```python
from os import name
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#Important for calculate the cross-correlation
from scipy.signal import correlate
from itertools import combinations

# Set names columns
# Date;Time;CO(GT);PT08.S1(CO);NMHC(GT);C6H6(GT);PT08.S2(NMHC);NOx(GT);PT08.S3(NOx);NO2(GT);PT08.S4(NO2);PT08.S5(O3);T;RH;AH;;
columns = ['Date', 'Time', 'CO(GT)', 'PT08.S1(CO)', 'NMHC(GT)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH', 'AH']
# Read csv file
dataset = pd.read_excel('dataset/AirQualityUCI.xlsx', skiprows=0, names=columns)
```

# Interpolate missing values

This function takes the data frame, **deletes** the NMHC(GT) column, and calculates the interpolation.

```python
def interpolate_missing_values(df):
    # Copy DataFrame
    df_interpolated = df.copy()
    
    # Delete NMHC(GT) column if exists
    if 'NMHC(GT)' in df_interpolated.columns:
        df_interpolated.drop('NMHC(GT)', axis=1, inplace=True)
    
    # Replace all missing values with NaN for prepare the interpolation
    df_interpolated.replace(-200, np.nan, inplace=True)
    
    # We need call infer_objects method, for use interpolation
    # https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.infer_objects.html
    #    Attempts soft conversion of object-dtyped columns, leaving non-object and unconvertible columns unchanged.
    df_interpolated = df_interpolated.infer_objects()

    # Execute linear interpolation
    # https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.interpolate.html
    #    Fill NaN values using an interpolation method.
    df_interpolated.interpolate(method='linear', limit_direction='forward', axis=0, inplace=True)
    
    # Correct NaN values with previous and next element of NaN
    df_interpolated.ffill(inplace=True)
    df_interpolated.bfill(inplace=True)
    # Return corrected dataset
    return df_interpolated
```

```python
dataset_interpolated = interpolate_missing_values(dataset)
```

# Cross-correlation

## Cross-correlation combinations

Then I've got the copy of the dataset, and I will work for the cross-correlation on these columns:
    
    ['CO(GT)' , 'PT08.S1(CO)' , 'C6H6(GT)' , 'PT08.S2(NMHC)' , 'NOx(GT)' , 'PT08.S3(NOx)' , 'NO2(GT)' , 'PT08.S4(NO2)' , 'PT08.S5(O3)' , 'T' , 'RH' , 'AH']

In total, we will have **66** combinations.

```python
#CO(GT);PT08.S1(CO);C6H6(GT);PT08.S2(NMHC);NOx(GT);PT08.S3(NOx);NO2(GT);PT08.S4(NO2);PT08.S5(O3);T;RH;AH;;
columns = ['CO(GT)' , 'PT08.S1(CO)' , 'C6H6(GT)' , 'PT08.S2(NMHC)' , 'NOx(GT)' , 'PT08.S3(NOx)' , 'NO2(GT)' , 'PT08.S4(NO2)' , 'PT08.S5(O3)' , 'T' , 'RH' , 'AH']
#columns = ['CO(GT)', 'C6H6(GT)', 'NOx(GT)', 'NO2(GT)']
df = dataset_interpolated.copy()
# Number index (i)
number = 1
# Total number of combinations
total = len(list(combinations(columns, 2)))
```

## Cross-correlation computing and results

```python
# Computing cross correlation between groups
for i in range(len(columns)):
    for j in range(i+1, len(columns)):
        col1, col2 = columns[i], columns[j]

        # Normalize the data
        series1 = (df[col1] - df[col1].mean()) / df[col1].std()
        series2 = (df[col2] - df[col2].mean()) / df[col2].std()

        # Using scipy, calculate the correlation between two series
        # https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.correlate.html
        #    Cross-correlate two N-dimensional arrays.
        #    Cross-correlate in1 and in2, with the output size determined by the mode argument. 
        #    full: The output is the full discrete linear cross-correlation of the inputs.
        cc = correlate(series1, series2, mode='full')

        # Normalize the cross-correlation by dividing the product of the series lengths
        cc /= np.sqrt(len(series1) * len(series2))

        # Take all lag valus with arange between series1 and series2
        # https://numpy.org/doc/stable/reference/generated/numpy.arange.html
        #    Return evenly spaced values within a given interval.
        lag = np.arange(-(len(series1)-1), len(series1))

        print(f"Plot Number[{number}/{total}]")
        print(f"Cross-correlation between {col1} e {col2}:")

        # Add number for counting every plot
        number += 1

        # Plot cross-correlation
        plt.figure(figsize=(10, 5))
        plt.plot(lag, cc)
        plt.title(f'Normalized Cross-Correlation between {col1} and {col2}')
        plt.xlabel('Lag')
        plt.ylabel('Normalized cross-correlation')
        plt.ylim(-1, 1)  # Set limit for y [1,-1]
        plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
        plt.show()
```
